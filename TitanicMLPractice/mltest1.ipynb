```python
#how to add columns do your data frame
my_df3 = my_df2.loc[:4]
#add column from a list (useful for small data frames, which will never happen haha...)
eye_color = ['green', 'brown', 'brown', 'blue', 'brown']
my_df3['eye color'] = eye_color
my_df3
#ignore the warning as thing is a odified trainingexample explaining a siple use of a pd function
```

```python
##add default values to data frame
#morbid example but very true for our data frame.
my_df3["Alive"] = [False]*len(my_df3)
my_df3
```

```python
#How to add a null or NaN value? use the same method above
my_df3["Birthday"] = [np.nan] * len(my_df3)
my_df3
```

```python
#the insert function should be WAYY more useful, honetly should have put this one higher up
#Use .insert() to add column to specific position
my_df3.insert(1, "Liked Boats", True)
my_df3
```

```python
#add columns with the .assign() function - this creates a new data frame so you don't accidentally destroy your reference data frame
#now we make a new dataset making sure we don't have any dogs messing up or data
my_df4 = my_df3.assign(Dog=[False]*len(my_df3))
my_df4
```

```python
#use the .drop() function to remove columns. let remove the "Birdthday" column i put there on "Purpose"
#make sure you use axis=1 for 2nd arg
#and also use inplace=True for the 3rd arg or pandas won't believe you
my_df4.drop("Birdthday", axis=1, inplace=True)
my_df4

#this kind of removed the data, but just imagine it totally did. We don't need to worry about that right now
```
